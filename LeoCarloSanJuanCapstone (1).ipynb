````python
# Filename: employability_model.py
# Run in a notebook or as a script (but plotting works best in a notebook)

import os
import pandas as pd
import numpy as np

# ML libraries
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import (
    roc_auc_score, classification_report, confusion_matrix,
    precision_recall_curve, average_precision_score, accuracy_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

# Explainability
import shap

# Plotting
import matplotlib.pyplot as plt

# -------------------------
# 1) Load data
# -------------------------
# DATA_PATH = "students_employability_ph.csv"  # change if needed
# df = pd.read_csv(DATA_PATH)

# Using the df already loaded from previous steps
df = df.copy() # Create a copy to avoid modifying the original df from previous cells

print("Rows,Cols:", df.shape)
display(df.head())

# -------------------------
# 2) Target selection
# -------------------------
# The dataset may have either a binary 'employable' column or a continuous 'employability_score'.
# Adjust according to your CSV's columns.

# The loaded dataset has a 'CLASS' column with 'Employable' and 'LessEmployable' values.
# Convert it to a binary 'employable' column (1 for Employable, 0 for LessEmployable).
if 'CLASS' in df.columns:
    target_col = 'employable'
    df[target_col] = df['CLASS'].apply(lambda x: 1 if x == 'Employable' else 0)
    df = df.drop(columns=['CLASS']) # Drop the original 'CLASS' column
elif 'employable' in df.columns:
    target_col = 'employable'            # already binary 0/1
elif 'employability_score' in df.columns:
    # convert to binary using threshold (e.g., top 60% considered employable)
    target_col = 'employable'
    df[target_col] = (df['employability_score'] >= df['employability_score'].median()).astype(int)
else:
    raise ValueError("Please ensure dataset has 'CLASS', 'employable' or 'employability_score' column.")

# -------------------------
# 3) Quick EDA (summary)
# -------------------------
print(df[target_col].value_counts(normalize=True))
print(df.describe(include='all').T.sample(10))

# -------------------------
# 4) Feature selection / engineering
# -------------------------
# Common features present in the employability dataset might include:
# - demographics: age, gender
# - academics: cgpa, grades, major
# - soft skills: communication_score, teamwork_score, confidence_score
# - technical skills flags: has_cert_autocad, has_cert_matlab, has_project
# - interview ratings: technical_rating, hr_rating, overall_mock_score
# - others: ojt_performance_score, hours_otj, club_participation

# Make a copy and drop obvious id columns
df_model = df.copy()
possible_id_cols = ['id', 'student_id', 'Name of Student'] # Added 'Name of Student' as a potential ID column
for c in possible_id_cols:
    if c in df_model.columns:
        df_model = df_model.drop(columns=[c])

# Fill in a list of candidate feature columns by excluding the target and any textual/irrelevant columns
exclude = [target_col]
cols = [c for c in df_model.columns if c not in exclude]
# If there are text fields (essays), consider NLP separately -- for now drop long text columns
text_cols = [c for c in cols if df_model[c].dtype == 'object' and df_model[c].str.len().fillna(0).mean() > 50]
if text_cols:
    print("Dropping long text columns for baseline:", text_cols)
    df_model = df_model.drop(columns=text_cols)
    cols = [c for c in df_model.columns if c not in exclude]

# Split numerical and categorical
num_cols = [c for c in cols if pd.api.types.is_numeric_dtype(df_model[c])]
cat_cols = [c for c in cols if c not in num_cols]

print("Numeric cols:", len(num_cols), num_cols)
print("Categorical cols:", len(cat_cols), cat_cols)

# -------------------------
# 5) Train/test split
# -------------------------
X = df_model[ num_cols + cat_cols ]
y = df_model[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

# -------------------------
# 6) Preprocessing pipeline
# -------------------------
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # Changed sparse to sparse_output
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ], remainder='drop'
)

# -------------------------
# 7) Build models
# -------------------------
# Logistic baseline
log_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42))
])

# Random Forest
rf_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', RandomForestClassifier(n_estimators=300, class_weight='balanced', random_state=42, n_jobs=-1))
])

# XGBoost
xgb_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', xgb.XGBClassifier(
        n_estimators=300, max_depth=6, learning_rate=0.05, use_label_encoder=False,
        eval_metric='logloss', scale_pos_weight = (y_train==0).sum() / (y_train==1).sum(),
        random_state=42
    ))
])

# -------------------------
# 8) Cross-validated scoring
# -------------------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for name, pipe in [('Logistic', log_pipe), ('RandomForest', rf_pipe), ('XGBoost', xgb_pipe)]:
    scores = cross_val_score(pipe, X_train, y_train, cv=skf, scoring='average_precision', n_jobs=-1)
    print(f"{name} AP (5-fold): mean={scores.mean():.3f} std={scores.std():.3f}")

# -------------------------
# 9) Fit best model and evaluate on test set
# -------------------------
best_pipe = xgb_pipe  # choose the best one from CV above; replace if RF/logistic is better
best_pipe.fit(X_train, y_train)

y_pred_proba = best_pipe.predict_proba(X_test)[:,1]
y_pred = (y_pred_proba >= 0.5).astype(int)

print("Test ROC AUC:", roc_auc_score(y_test, y_pred_proba))
print("Average Precision (PR AUC):", average_precision_score(y_test, y_pred_proba))
print(classification_report(y_test, y_pred))

# Confusion matrix (useful to pick threshold)
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)

# -------------------------
# 10) Precision-Recall curve & choose threshold
# -------------------------
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)
ap = average_precision_score(y_test, y_pred_proba)
plt.figure(figsize=(6,4))
plt.plot(recall, precision, label=f'AP={ap:.3f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()

# Choose threshold to get high recall (catch at-risk students)
# Example: pick threshold achieving recall >= 0.80 on test set
target_recall = 0.80
thr = 0.5
for t in np.linspace(0.01, 0.99, 99):
    yhat = (y_pred_proba >= t).astype(int)
    r = ( (y_test==1) & (yhat==1) ).sum() / (y_test==1).sum()
    if r >= target_recall:
        thr = t
        break
print("Selected threshold for recall>=0.8:", thr)

# -------------------------
# 11) Explainability with SHAP (TreeExplainer for XGBoost)
# -------------------------
# Note: we must recreate feature names after preprocessing for meaningful SHAP
preproc = preprocessor.fit(X_train)  # ensure fitted
X_train_tr = preproc.transform(X_train)

# Build feature names: numeric + onehot categories
ohe = preproc.named_transformers_['cat'].named_steps['onehot']
cat_feature_names = []
if len(cat_cols) > 0:
    cat_feat_vals = ohe.get_feature_names_out(cat_cols)
    cat_feature_names = list(cat_feat_vals)
feature_names = list(num_cols) + cat_feature_names

# Use SHAP
model = best_pipe.named_steps['clf']
explainer = shap.TreeExplainer(model)
# pass dense array or DataFrame with feature_names
X_test_tr = preproc.transform(X_test)
shap_values = explainer.shap_values(X_test_tr)

# Summary plot (global)
shap.summary_plot(shap_values, features=X_test_tr, feature_names=feature_names, show=True)

# Local explanation example - print top shap features for a single student
i = 0
sv = shap_values[i]
# show top 8 features by absolute shap value
top_idx = np.argsort(np.abs(sv))[-8:]
print("Top contributing features for example student:", np.array(feature_names)[top_idx])
print("SHAP values:", sv[top_idx])
````

## Define and Configure Supervised Models

### Subtask:
Set up pipelines for Logistic Regression, Decision Trees, Random Forest, XGBoost, and Support Vector Machines (SVM). Each pipeline will include the preprocessor followed by the respective classifier. I will also include suitable hyperparameters and ensure class imbalance is handled where appropriate.


**Reasoning**:
I will import the necessary libraries, calculate the class imbalance weight for XGBoost, and then define and configure pipelines for Logistic Regression, Decision Tree, Random Forest, XGBoost, and SVM, incorporating the `preprocessor` and specified hyperparameters for each.



````python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

# Calculate scale_pos_weight for handling class imbalance in XGBoost
# This is the ratio of negative samples to positive samples.
scale_pos_weight_value = (y_train == 0).sum() / (y_train == 1).sum()

# 1. Logistic Regression pipeline
log_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42))
])

# 2. Decision Tree pipeline
dt_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', DecisionTreeClassifier(class_weight='balanced', random_state=42, max_depth=10))
])

# 3. Random Forest pipeline
rf_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', RandomForestClassifier(n_estimators=300, class_weight='balanced', random_state=42, n_jobs=-1))
])

# 4. XGBoost pipeline
xgb_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', xgb.XGBClassifier(
        n_estimators=300, max_depth=6, learning_rate=0.05, use_label_encoder=False,
        eval_metric='logloss', scale_pos_weight=scale_pos_weight_value,
        random_state=42
    ))
])

# 5. Support Vector Machine (SVM) pipeline
svm_pipe = Pipeline(steps=[
    ('preproc', preprocessor),
    ('clf', SVC(probability=True, class_weight='balanced', random_state=42))
])

print("All classifier pipelines (Logistic Regression, Decision Tree, Random Forest, XGBoost, SVM) have been defined.")
````

USING THE DIFFERENT MODELS TO EVALUATE PERFORMANCE


**Reasoning**:
Now that the pipelines are defined, I will create a list of these pipelines to iterate through them. For each pipeline, I will train the model on the training data, record the training time, make predictions on the test data, and calculate various evaluation metrics (Accuracy, Precision, Recall, F1-Score, ROC AUC).



````python
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# List of pipelines to evaluate
pipelines = {
    'Logistic Regression': log_pipe,
    'Decision Tree': dt_pipe,
    'Random Forest': rf_pipe,
    'XGBoost': xgb_pipe,
    'SVM': svm_pipe
}

# Dictionary to store results
results = {
    'Model': [],
    'Training Time (s)': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1-Score': [],
    'ROC AUC': []
}

# Iterate through each pipeline, train, predict, and evaluate
for model_name, pipeline in pipelines.items():
    print(f"\n--- Training and evaluating {model_name} ---")

    start_time = time.time()
    pipeline.fit(X_train, y_train)
    training_time = time.time() - start_time

    y_pred = pipeline.predict(X_test)

    # For ROC AUC, we need probabilities. SVC needs probability=True to enable predict_proba
    if hasattr(pipeline, 'predict_proba'):
        y_pred_proba = pipeline.predict_proba(X_test)[:, 1]
    else:
        # For models without predict_proba (like some SVMs), this will raise an error
        # Or, if decision_function is available, convert to probabilities
        y_pred_proba = pipeline.decision_function(X_test)
        # Scale decision_function output to probabilities if needed, for simplicity we skip it for now if not directly proba
        # This might not be ideal for SVM if not calibrated, but for this step we proceed as per instruction to get a value

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba) if 'y_pred_proba' in locals() and y_pred_proba.ndim > 0 else None

    results['Model'].append(model_name)
    results['Training Time (s)'].append(training_time)
    results['Accuracy'].append(accuracy)
    results['Precision'].append(precision)
    results['Recall'].append(recall)
    results['F1-Score'].append(f1)
    results['ROC AUC'].append(roc_auc)

    print(f"Training Time: {training_time:.2f} seconds")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")

# Create a DataFrame from the results
performance_df = pd.DataFrame(results)

print("\nModel performance results have been calculated.")
````

## Prepare Data for Deep Learning

### Subtask:
Convert the preprocessed X_train, X_test, y_train, and y_test data into PyTorch tensors suitable for a deep learning model.


**Reasoning**:
I will import the `torch` library, then apply the preprocessor to `X_train` and `X_test` and convert them to `float32` PyTorch tensors. Similarly, I will convert `y_train` and `y_test` to `float32` PyTorch tensors and reshape them to column vectors. Finally, I will print the shapes of all resulting tensors to verify their dimensions.



````python
import torch

# 1. Apply preprocessor to X_train and convert to PyTorch tensor
X_train_processed = preprocessor.transform(X_train)
X_train_tensor = torch.tensor(X_train_processed, dtype=torch.float32)

# 2. Apply preprocessor to X_test and convert to PyTorch tensor
X_test_processed = preprocessor.transform(X_test)
X_test_tensor = torch.tensor(X_test_processed, dtype=torch.float32)

# 3. Convert y_train to PyTorch tensor and reshape
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).reshape(-1, 1)

# 4. Convert y_test to PyTorch tensor and reshape
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).reshape(-1, 1)

# 5. Print the shapes of the created tensors
print(f"Shape of X_train_tensor: {X_train_tensor.shape}")
print(f"Shape of X_test_tensor: {X_test_tensor.shape}")
print(f"Shape of y_train_tensor: {y_train_tensor.shape}")
print(f"Shape of y_test_tensor: {y_test_tensor.shape}")
````

## Define Deep Learning Model

### Subtask:
Define a simple neural network architecture using PyTorch. This will include input, hidden, and output layers, with appropriate activation functions. Consider the number of features in the preprocessed data for the input layer and the binary nature of the target for the output layer.


**Reasoning**:
To define the neural network architecture, I need to import the `torch.nn` module as `nn` and `torch.nn.functional` as `F`. Then, I will create a Python class `BinaryClassifier` that inherits from `nn.Module` and define the input, hidden, and output layers with appropriate activation functions for binary classification, as specified in the instructions.



````python
import torch.nn as nn
import torch.nn.functional as F

# Define the neural network architecture
class BinaryClassifier(nn.Module):
    def __init__(self, input_features):
        super(BinaryClassifier, self).__init__()
        self.hidden_layer_1 = nn.Linear(input_features, 64)
        self.hidden_layer_2 = nn.Linear(64, 32)
        self.output_layer = nn.Linear(32, 1)

    def forward(self, x):
        x = F.relu(self.hidden_layer_1(x))
        x = F.relu(self.hidden_layer_2(x))
        x = torch.sigmoid(self.output_layer(x)) # Sigmoid for binary classification output
        return x

print("Neural network architecture defined.")

# Instantiate the model (using the shape of X_train_tensor from previous step)
input_features = X_train_tensor.shape[1]
model = BinaryClassifier(input_features)
print(f"Model instantiated with {input_features} input features.")
print(model)
````

## Train Deep Learning Model

### Subtask:
Train the defined deep learning model on the prepared training data. This will involve setting up a loss function (e.g., Binary Cross-Entropy) and an optimizer (e.g., Adam). Track the training time for comparison.


**Reasoning**:
I will define the loss function, optimizer, and number of epochs, then train the deep learning model using a loop that performs forward pass, calculates loss, backpropagation, and updates weights, while also tracking the training time.



````python
import torch.optim as optim
import time

# 1. Define the loss function and optimizer
criterion = nn.BCELoss() # Binary Cross-Entropy Loss for binary classification
optimizer = optim.Adam(model.parameters(), lr=0.001) # Adam optimizer with learning rate 0.001

# 2. Set the number of training epochs
epochs = 100

# 3. Initialize a variable to record the start time of the training
start_time = time.time()

# 4. Loop through the specified number of epochs
print("Starting model training...")
for epoch in range(epochs):
    # a. Perform a forward pass
    outputs = model(X_train_tensor)

    # b. Calculate the loss
    loss = criterion(outputs, y_train_tensor)

    # c. Clear the gradients
    optimizer.zero_grad()

    # d. Perform backpropagation
    loss.backward()

    # e. Update the model's weights
    optimizer.step()

    # f. Print the loss every few epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# 5. Calculate the total training time
training_time = time.time() - start_time

# 6. Print a confirmation message and display the total training time
print("Model training complete.")
print(f"Total training time: {training_time:.2f} seconds")
````

## Evaluate Deep Learning Model

### Subtask:
Evaluate the trained deep learning model on the prepared test data. Calculate Accuracy, Precision, Recall, F1-Score, and ROC AUC for the deep learning model.


**Reasoning**:
To evaluate the deep learning model, I will use the trained model to make predictions on the test data, convert these predictions to binary outputs, and then calculate and print the specified evaluation metrics (Accuracy, Precision, Recall, F1-Score, and ROC AUC).



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 1. Set the model to evaluation mode and disable gradient calculation
model.eval()
with torch.no_grad():
    # Make predictions on the test data
    y_pred_proba_dl = model(X_test_tensor)

# 2. Convert continuous predictions (probabilities) to binary predictions (0 or 1)
y_pred_dl = (y_pred_proba_dl >= 0.5).float()

# 3. Convert y_test_tensor and binary predictions to NumPy arrays
y_test_np = y_test_tensor.numpy()
y_pred_np = y_pred_dl.numpy()
y_pred_proba_np = y_pred_proba_dl.numpy()

# 4. Calculate evaluation metrics
accuracy_dl = accuracy_score(y_test_np, y_pred_np)
precision_dl = precision_score(y_test_np, y_pred_np)
recall_dl = recall_score(y_test_np, y_pred_np)
f1_dl = f1_score(y_test_np, y_pred_np)
roc_auc_dl = roc_auc_score(y_test_np, y_pred_proba_np)

# 5. Print each calculated metric
print("\n--- Deep Learning Model Evaluation ---")
print(f"Accuracy: {accuracy_dl:.4f}")
print(f"Precision: {precision_dl:.4f}")
print(f"Recall: {recall_dl:.4f}")
print(f"F1-Score: {f1_dl:.4f}")
print(f"ROC AUC: {roc_auc_dl:.4f}")
````

### Explanation of SHAP Values

SHAP (SHapley Additive exPlanations) values are a powerful tool for interpreting machine learning models. They quantify the contribution of each feature to the prediction of an individual instance, explaining how each feature pushes the model's output from the base value (expected output of the model) to the actual output.

**Global Feature Importance (SHAP Summary Plot):**

The SHAP summary plot provides a global overview of feature importance. Each point on the plot represents a Shapley value for a feature and an instance. The plot shows:

*   **Feature Importance**: Features are ranked by the absolute average of their SHAP values, indicating which features have the largest impact on model predictions overall.
*   **Direction of Impact**: The color of each point (e.g., red for high feature value, blue for low feature value) indicates whether that feature value tends to push the prediction higher or lower. For example, if high values of 'Student Performance Rating' (red dots) are concentrated on the positive side of the plot, it means higher performance ratings generally increase the likelihood of being classified as 'Employable'.
*   **Distribution**: The spread of the points along the x-axis shows the range of impact each feature has.

**Local Explanation (SHAP Force Plot):**

For a specific instance, the SHAP force plot visualizes how each feature contributes to that single prediction. It breaks down the prediction for an individual instance by showing:

*   **Base Value (E[f(x)])**: This is the average output of the model over the entire dataset.
*   **Features Pushing Up**: Features that increase the prediction from the base value are shown in red, pushing the prediction to the right.
*   **Features Pushing Down**: Features that decrease the prediction from the base value are shown in blue, pushing the prediction to the left.
*   **Magnitude of Contribution**: The length of each bar indicates the magnitude of the feature's influence. Larger bars signify a greater impact.

In essence, SHAP values help us understand both *which* features are important and *how* they influence predictions, both globally across the dataset and locally for individual predictions. For our best model, XGBoost, the SHAP analysis helps us understand why a student is predicted as 'Employable' or 'LessEmployable' based on their attributes.

### Explanation of Fairness Metrics

#### 1. Demographic Parity

*   **What it is**: Demographic Parity (also known as Statistical Parity) is a fairness metric that requires a model's prediction of a positive outcome to be independent of the sensitive attribute. In simpler terms, it means that the proportion of individuals receiving a positive outcome (e.g., being classified as 'Employable') should be roughly equal across different groups defined by a sensitive attribute (e.g., gender, race).
*   **How it measures fairness**: It measures fairness by checking if the positive prediction rate P(Y_hat=1) is the same for all sensitive groups A=a. For instance, P(Y_hat=1 | Gender=Female) should be approximately equal to P(Y_hat=1 | Gender=Male).
*   **Application with sensitive attributes**: If sensitive attributes like 'Gender' or 'Ethnicity' were available in our dataset, we would calculate the percentage of 'Employable' predictions for each demographic group. For example, we'd compare the employability rate predicted by the model for female students versus male students, or for students of different ethnic backgrounds. A significant difference would indicate a violation of demographic parity.

#### 2. Equal Opportunity

*   **What it is**: Equal Opportunity is a fairness metric that focuses on ensuring that groups receive a fair chance at a positive outcome when they are truly deserving of it. Specifically, it requires that the True Positive Rate (TPR)  the proportion of actual positive cases that are correctly identified  is equal across different groups.
*   **How it measures fairness**: It measures fairness by checking if the true positive rate P(Y_hat=1 | Y=1) is the same for all sensitive groups A=a. For instance, P(Y_hat=1 | Y=1, Gender=Female) should be approximately equal to P(Y_hat=1 | Y=1, Gender=Male). This means that among truly employable students, the model should correctly identify them at the same rate regardless of their gender.
*   **Application with sensitive attributes**: If 'Gender' or 'Ethnicity' were available, we would calculate the True Positive Rate (Recall) for each group. For example, among students who *are* truly employable, we would compare the model's ability to correctly classify them as 'Employable' for female students versus male students. Disparities in TPR would suggest that one group has less equal opportunity to be correctly identified as employable by the model.

#### 3. Predictive Equality

*   **What it is**: Predictive Equality (also known as False Positive Rate Parity) is a fairness metric that requires that the False Positive Rate (FPR)  the proportion of actual negative cases that are incorrectly identified as positive  is equal across different groups.
*   **How it measures fairness**: It measures fairness by checking if the false positive rate P(Y_hat=1 | Y=0) is the same for all sensitive groups A=a. For instance, P(Y_hat=1 | Y=0, Gender=Female) should be approximately equal to P(Y_hat=1 | Y=0, Gender=Male). This implies that the model should incorrectly flag non-employable students as 'Employable' at the same rate across different genders.
*   **Application with sensitive attributes**: With 'Gender' or 'Ethnicity' data, we would calculate the False Positive Rate for each group. For example, among students who *are not* truly employable, we would compare how often the model mistakenly predicts them as 'Employable' for female students versus male students. A higher FPR for one group would mean they are unfairly disadvantaged by being incorrectly labeled as employable, which could lead to wasted resources or false hopes.

### Proposed Methods to Audit for Potential Indirect Biases

To proactively identify and mitigate these indirect biases, several auditing methods can be employed:

a.  **Statistical Analysis**: This involves looking for statistically significant differences in feature ratings or model outcomes across inferred or proxy groups. For example:
    *   **Distributional Analysis**: Examine the distribution of 'GENERAL APPEARANCE' ratings. Are certain ranges of scores disproportionately low for individuals whose names might suggest a particular ethnicity (a proxy group, if ethically justifiable for analysis)?
    *   **Correlation with Outcomes**: Analyze if specific subjective features correlate strongly with the 'employable' outcome in ways that cannot be explained by job-relevant skills, and then investigate the characteristics of individuals receiving high/low scores on these features.
    *   **Disparate Impact**: Test for disparate impact, where a seemingly neutral feature disproportionately affects a protected group, even if the group is only inferred.

b.  **Human Auditing / Qualitative Studies**: Directly engage with human evaluators to understand their rating processes and potential blind spots:
    *   **Interviewer Debriefs**: Conduct structured interviews or focus groups with the individuals who provided these ratings to understand their criteria, interpretations, and any conscious or unconscious biases they might hold.
    *   **Bias Training Effectiveness**: Assess if bias training for raters changes the distribution or correlation of these subjective scores.
    *   **Rubric Review**: Critically evaluate the rubrics or guidelines used for rating to identify vague language or criteria that are susceptible to subjective interpretation and bias.

c.  **External Data Comparison (Ethical & Permissible)**: If comparable external datasets with demographic information exist and can be ethically used, cross-reference patterns:
    *   **Benchmarking**: Compare the distribution of subjective feature scores in our dataset with distributions from external, unbiased benchmarks (if available) to detect deviations.
    *   **Demographic Skew**: If demographic data is collected for other purposes (e.g., diversity reporting post-hiring), compare the 'GENERAL APPEARANCE' scores of hired individuals against those not hired, stratified by demographic groups.

d.  **Feature Perturbation/Ablation**: This method involves systematically testing the model's sensitivity to changes in these features:
    *   **Sensitivity Testing**: Systematically vary the values of 'GENERAL APPEARANCE' for a subset of individuals while keeping other features constant and observe how the model's 'employable' prediction changes. If a minor, non-job-relevant change in appearance drastically alters the prediction, it indicates over-reliance or bias.
    *   **Ablation Studies**: Train the model with and without these subjective features to quantify their overall impact on predictive performance and fairness metrics. If removing them improves fairness without significantly degrading performance, it suggests they were contributing to bias.

### Challenges and Ethical Considerations

Auditing for indirect biases without direct sensitive attributes presents significant challenges:

*   **Proxy Identification**: Accurately identifying proxy groups (e.g., inferring ethnicity from names or zip codes) can be fraught with error and introduce new biases, as proxies are often imperfect and can reinforce stereotypes.
*   **Ethical Dilemma**: Using proxies raises ethical concerns about privacy and potentially using information to re-create a discriminatory system, even if the intent is to mitigate bias.
*   **Causality vs. Correlation**: It's challenging to disentangle whether a correlation between a subjective feature and an outcome is due to legitimate job requirements or an indirect bias. For example, 'MANNER OF SPEAKING' might be genuinely important for a sales role, but biases can still exist.
*   **Lack of Ground Truth**: Without explicit sensitive attributes, defining and measuring 'fairness' becomes complex. There's no clear ground truth for what an unbiased outcome *should* look like across various demographic dimensions.
*   **Interpretability**: Complex models can make it difficult to attribute specific biases to particular features. SHAP values help, but the interpretation of 'bias' from feature contributions still requires domain expertise and ethical judgment.
*   **Resource Intensive**: Implementing robust auditing methods, especially human and qualitative studies, can be time-consuming and expensive.

Addressing indirect bias requires a multi-faceted approach, combining statistical rigor with qualitative insights and continuous ethical reflection, even when direct demographic data is not available.

    *   **Predictive Equality:** Requires that the False Positive Rate (proportion of actual negative cases incorrectly identified as positive) is equal across different groups.

4.  **How would these fairness metrics be applied if sensitive attributes were available in this dataset?**
    *   **Demographic Parity:** One would compare the percentage of 'Employable' predictions for each demographic group (e.g., male vs. female students).
    *   **Equal Opportunity:** One would calculate and compare the True Positive Rate (Recall) for each sensitive group among those who are truly employable.
    *   **Predictive Equality:** One would calculate and compare the False Positive Rate for each sensitive group among those who are truly not employable.

5.  **What are the limitations of applying these fairness metrics to the current dataset?**
    The complete absence of sensitive attributes in the dataset makes it impossible to directly compute or apply Demographic Parity, Equal Opportunity, or Predictive Equality.

6.  **Can features like 'GENERAL APPEARANCE' or 'MANNER OF SPEAKING' introduce bias, even without explicit demographic data?**
    Yes, features such as 'GENERAL APPEARANCE', 'MANNER OF SPEAKING', and 'PHYSICAL CONDITION' are highly subjective and can easily introduce implicit biases. They can serve as proxies for protected attributes or perpetuate societal stereotypes (e.g., appearance influenced by socioeconomic status, speaking manner by region/ethnicity), leading to unfair outcomes.

7.  **What methods are proposed to audit these subjective features for potential indirect biases?**
    Proposed auditing methods include:
    *   **Statistical Analysis:** Examining feature distributions, correlations with outcomes, and disparate impact across inferred or proxy groups.
    *   **Human Auditing/Qualitative Studies:** Engaging with human evaluators to understand their rating processes, biases, and reviewing rubrics.
    *   **External Data Comparison:** (If ethically permissible) Benchmarking subjective feature scores against external demographic distributions.
    *   **Feature Perturbation/Ablation:** Systematically varying or removing these features to test the model's sensitivity and impact on predictions and fairness.
